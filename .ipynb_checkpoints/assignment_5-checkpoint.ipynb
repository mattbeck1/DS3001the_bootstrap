**Q2.** This question refers to the `mammogram.csv` data. It has two variables, `treatment` which takes the values `control` or `mammogram`, and `breast_cancer_death`, which takes the values `no` or `yes`. This is an experiment that followed 89,835 women for 25 years to see if mammograms were superior to more traditional breast cancer screenings in preventing breast cancer deaths.

```python
# Imports
import pandas as pd
import numpy as np
import math
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Data
df = pd.read_csv('./data/mammogram.csv')
```

```python
df.head()
```

1. Cross tabulate `treatment` and `breast_cancer_death`. What is the difference in 25-year survival rates between the control and mammogram groups?

```python
crosstable = pd.crosstab(df['treatment'], df['breast_cancer_death'])
print(crosstable)
```

```python
total_control = sum(crosstable.loc['control'])
total_mammogram = sum(crosstable.loc['mammogram'])

rate_control = crosstable.loc['control', 'no']/total_control
rate_mammogram = crosstable.loc['mammogram', 'no']/total_mammogram

print("Survival Rate (Control): ", rate_control)
print("Survival Rate (Mammogram): ", rate_mammogram)

diff_0 = rate_control - rate_mammogram

print("Difference between Control and Mammogram Survival Rates: ", diff_0)
```

2. Bootstrap the densities and distributions of survival rates for the two groups. 

```python
df['survive'] = 0
df.loc[df['breast_cancer_death']=='no', 'survive']=1
# Control Data
df_control = df.loc[df['treatment'] == 'control', 'survive']
N_control = df_control.shape[0]
# Mammogram Data
df_mammogram = df.loc[df['treatment'] == 'mammogram', 'survive']
N_mammogram = df_mammogram.shape[0]
```

```python
# Simulation
S = 1000
rate_C = np.zeros(S)
rate_M = np.zeros(S)
diff = np.zeros(S)

# Bootstrapping
for s in range(S):
    sample_C = df_control.sample(N_control, axis = 0, replace = True)
    rate_C[s] = sample_C.mean()
    sample_M = df_mammogram.sample(N_mammogram, axis = 0, replace = True)
    rate_M[s] = sample_M.mean()
    diff[s] = rate_C[s] - rate_M[s]
    
# Make dataframes
rdf_C = pd.DataFrame({'treatment':'control', 'survival_rate':rate_C})
rdf_M = pd.DataFrame({'treatment':'mammogram', 'survival_rate':rate_M})
rdf = pd.concat([rdf_C, rdf_M], axis = 0)
```

```python
plt.figure(figsize=(10,6))
sns.kdeplot(data=rdf, x='survival_rate', hue='treatment')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.ecdfplot(data=rdf, x='survival_rate', hue='treatment')
plt.show()
```

3. Construct a 99% confidence interval for the difference in outcomes bewteen the two groups. Does it include zero?

```python
CI = np.quantile(diff, [0.005, 0.995])
print('CI: ' + str(CI))
```

```python
plt.figure(figsize=(10,6))
ax = sns.kdeplot(diff)
ax.axvline(x = diff_0,linestyle='dashed',color='green')
ax.axvline(x = CI[0],linestyle='dashed',color='red')
ax.axvline(x = CI[1],linestyle='dashed',color='red')
plt.show()
```

**Q3.** This question refers to the `diabetes_hw.csv` data. It contains two variables, `outcome` and `treatment`. Each is looking at whether an individual's diabetes was successfully treated (`outcome==success`) with `lifestyle` interventions like exercises and diets, a drug denoted by `met` (metformin), or a drug denoted by `rosi` (rosiglitazone), or not (`outcome==failure`). Treatment success means that the individual no longer needs to be treated with insulin, while failure means the patient still required insulin injections after treatment.

```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
# Load the Data
df = pd.read_csv('./data/diabetes_hw.csv')
```

1. Cross tabulate `treatment` and `outcome`.

```python
crosstable = pd.crosstab(df['treatment'], df['outcome'])
print(crosstable)
```

2. Compute the the proportion of successes for each treatment. Which treatment appears to be the most effective?

```python
lifestyle_rate = crosstable.loc['lifestyle', 'success']/sum(crosstable.loc['lifestyle'])
print("Lifestlye Success Rate: ", lifestyle_rate)

met_rate = crosstable.loc['met', 'success']/sum(crosstable.loc['met'])
print("Met Success Rate: ", met_rate)

rosi_rate = crosstable.loc['rosi', 'success']/sum(crosstable.loc['rosi'])
print("Rosi Success Rate: ", rosi_rate)
```

Rosi has the highest success rate.

3. Bootstrap the density and distribution of the proportion of successes for each interventions. Create empirical CDF and kernel density plots that are grouped  by treatment type. Which treatment appears to be the most effective?

```python
df['success'] = 0
df.loc[df['outcome'] == 'success', 'success'] = 1
# Lifestyle Data
df_L = df.loc[df['treatment'] == 'lifestyle', 'success']
N_L = df_L.shape[0]
# Met Data
df_M = df.loc[df['treatment'] == 'met', 'success']
N_M = df_M.shape[0]
# Rosi Data
df_R = df.loc[df['treatment'] == 'rosi', 'success']
N_R = df_R.shape[0]
```

```python
# Simulation
S = 1000
rate_L = np.zeros(S)
rate_M = np.zeros(S)
rate_R = np.zeros(S)

# Bootstrapping
for s in range(S):
    sample_L = df_L.sample(N_L, axis = 0, replace = True)
    rate_L[s] = sample_L.mean()
    sample_M = df_M.sample(N_M, axis = 0, replace = True)
    rate_M[s] = sample_M.mean()
    sample_R = df_R.sample(N_R, axis = 0, replace = True)
    rate_R[s] = sample_R.mean()
    
# Make dataframes
rdf_L = pd.DataFrame({'treatment':'lifestyle', 'success_rate':rate_L})
rdf_M = pd.DataFrame({'treatment':'met', 'success_rate':rate_M})
rdf_R = pd.DataFrame({'treatment':'rosi', 'success_rate':rate_R})
rdf = pd.concat([rdf_L, rdf_M, rdf_R], axis = 0)
```

```python
plt.figure(figsize=(10,6))
sns.kdeplot(data=rdf, x='success_rate', hue='treatment')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.ecdfplot(data=rdf, x='success_rate', hue='treatment')
plt.show()
```

4. For each comparison (lifestyle versus met, met versus rosi, rosi versus lifestyle), bootstrap the distribution of the difference in outcomes. At the 90% level of confidence, which pairwise treatment comparisons are significantally different?

```python
# Simulation
S = 1000
rate_LvM = np.zeros(S)
rate_MvR = np.zeros(S)
rate_RvL = np.zeros(S)

# Bootstrapping
for s in range(S):
    sample_L = df_L.sample(N_L, axis = 0, replace = True)
    r_L = sample_L.mean()
    sample_M = df_M.sample(N_M, axis = 0, replace = True)
    r_M = sample_M.mean()
    sample_R = df_R.sample(N_R, axis = 0, replace = True)
    r_R = sample_R.mean()
    rate_LvM[s] = r_L - r_M
    rate_MvR[s] = r_M - r_R
    rate_RvL[s] = r_R - r_L
    
# Make dataframes
rdf_LvM = pd.DataFrame({'treatment':'lifestyle versus met', 'difference':rate_LvM})
rdf_MvR = pd.DataFrame({'treatment':'met versus rosi', 'difference':rate_MvR})
rdf_RvL = pd.DataFrame({'treatment':'rosi versus lifestyle', 'difference':rate_RvL})
rdf = pd.concat([rdf_LvM, rdf_MvR, rdf_RvL], axis = 0)
```

```python
plt.figure(figsize=(10,6))
sns.kdeplot(data=rdf, x='difference', hue='treatment')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.ecdfplot(data=rdf, x='difference', hue='treatment')
plt.show()
```

5. Which treatment appears to be the most effective overall?

The Rosi treatment appears to be the most effective overall. Through the bootstrapping we can see that the center of its distribution is the greatest. Additionally, the boostrapping performed on the differences between the treatments shows that rosi vs the alternatives has the largest differences. Overall, we can conclude that rosi has the highest success rate, making it the most effective.
